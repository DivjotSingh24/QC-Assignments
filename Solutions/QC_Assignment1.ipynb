```python
import numpy as np
import math
```

```python
a=1/math.sqrt(2)
b=1/math.sqrt(2)
v=np.array([a,b])
print(v)
```

```python
norm_check=a*a+b*b
print(norm_check)
if np.isclose(norm_check,1):
    print("Vector is normalized")
else:
    print("Vector is NOT normalized")
```

```python
length=np.linalg.norm(v)
print("Length of vector:",length)
```

```python
n=int(input("Enter dimension of vector: "))
A=[]
for i in range(n):
    A.append(float(input(f"Enter element {i+1}: ")))
A=np.array(A)
print("Vector A:", A)
```

```python
def check_normalized(v):
    return np.isclose(np.linalg.norm(v), 1)
print("Normalized:",check_normalized(A))
```

```python
B=np.array(list(map(float, input("Enter vector B: ").split())))
dot=np.dot(A, B)
cross=np.cross(A,B)
print("Dot product:",dot)
print("Cross product:",cross)
```

```python
def unit_vector(v):
    return v/np.linalg.norm(v)
print("Unit vector of A:", unit_vector(A))
```

```python
r=int(input("Rows: "))
c=int(input("Cols: "))

M=[]
for i in range(r):
    row=list(map(float,input().split()))
    M.append(row)
M=np.array(M)
print(M)
```

```python
print("Transpose:\n", M.T)
print("Determinant:", np.linalg.det(M))
print("Rank:", np.linalg.matrix_rank(M))
print("Trace:", np.trace(M))
```

```python
values,vectors=np.linalg.eig(M)
print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)

print("Inverse:\n", np.linalg.inv(M))
```
